### Slicing

```python
x = np.arange(10)
```

```python
x[:4]
```

```python
y = x[:4]
```

```python
y
```

```python
y[0] = 100
```

```python
y
```

```python
x
```

Notice that modifying y also modified x. That is because unlike Python lists, slicing numpy arrays doesn't create a new array. It just creates a new view into the existing array. 

The `copy` method should be used to make an explicit copy of an array.

```python
x = np.arange(10)
y = x[:4].copy()
y[0] = 100
print(y)
print(x)
```

### Sub Plots

```python
fig = plt.figure()

x = np.arange(50)
y = np.random.randn(50).cumsum()


# create 2x2 grid and this is number 1
ax1 = fig.add_subplot(2, 2, 1) 
ax1.plot(x, y)

# subplot 2
ax2 = fig.add_subplot(2, 2, 2) 
ax2.hist(y)

# subplot 3
ax2 = fig.add_subplot(2, 2, 3) 
ax2.scatter(x, y)
```

```python
x = np.arange(50)
y = np.random.randn(50).cumsum()

fig, plots = plt.subplots(2)

plots[0].plot(x, y)
plots[1].hist(y)
```
